```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
# Load Iris dataset
iris = load_iris()
iris
data = iris.data
target = iris.target
```

```python
# Standardize the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
# Visualize the data
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=target)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Dataset - Sepal Length vs Sepal Width')
plt.show()
```

```python
# K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data_scaled)

# Hierarchical clustering
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical.fit(data_scaled)
```

```python
# Get cluster centroids for K-means
kmeans_centroids_2d = scaler.inverse_transform(kmeans.cluster_centers_)[:, :2]

# Calculate cluster centroids for hierarchical clustering
hierarchical_centroids_2d = []
for cluster_label in np.unique(hierarchical.labels_):
    cluster_points = data[hierarchical.labels_ == cluster_label]
    cluster_centroid = np.mean(cluster_points[:, :2], axis=0)
    hierarchical_centroids_2d.append(cluster_centroid)

hierarchical_centroids_2d = np.array(hierarchical_centroids_2d)

print("K-means Cluster Centroids (2D):")
print(kmeans_centroids_2d)

print("\nHierarchical Cluster Centroids (2D):")
print(hierarchical_centroids_2d)
```

```python
# Visualize K-means clustering results (2D)
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_)
plt.scatter(kmeans_centroids_2d[:, 0], kmeans_centroids_2d[:, 1], c='red', marker='x', label='K-means Centroids')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-means Clustering (2D)')
plt.legend()
plt.show()

# Visualize hierarchical clustering results (2D)
plt.scatter(data[:, 0], data[:, 1], c=hierarchical.labels_)
plt.scatter(hierarchical_centroids_2d[:, 0], hierarchical_centroids_2d[:, 1], c='red', marker='x', label='Hierarchical Centroids')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Hierarchical Clustering (2D)')
plt.legend()
plt.show()
```

```python
# Perform PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(data_scaled)
print(pca_data)

# Plot PCA results
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=target)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Iris Dataset - PCA')
plt.show()
```
